```python
import cv2
import numpy as np
from keras.models import model_from_json
from keras import layers, models

# Load the model architecture
with open("facialemotionmodel.json", "r") as json_file:
    model_json = json_file.read()

# Load the model with custom objects
model = model_from_json(model_json, custom_objects={
    "Sequential": models.Sequential,
    "Conv2D": layers.Conv2D,
    "MaxPooling2D": layers.MaxPooling2D,
    "Dropout": layers.Dropout,
    "Flatten": layers.Flatten,
    "Dense": layers.Dense
})

# Load the model weights
model.load_weights("facialemotionmodel.h5")

# Load Haar Cascade for face detection
haar_file = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier(haar_file)

# Function to preprocess the face image
def extract_features(image):
    feature = np.array(image)
    feature = feature.reshape(1, 48, 48, 1)
    return feature / 255.0

# Open the webcam
webcam = cv2.VideoCapture(0)

# Emotion labels
labels = {0: 'angry', 1: 'disgust', 2: 'fear', 3: 'happy', 4: 'neutral', 5: 'sad', 6: 'surprise'}

while True:
    # Read a frame from the webcam
    ret, im = webcam.read()
    if not ret:
        break

    # Convert to grayscale
    gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

    # Detect faces in the image
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    # Process each detected face
    for (x, y, w, h) in faces:
        face_img = gray[y:y+h, x:x+w]  # Crop the face
        face_img = cv2.resize(face_img, (48, 48))  # Resize to model input size
        img = extract_features(face_img)  # Preprocess the image

        # Predict emotion
        pred = model.predict(img)
        prediction_label = labels[pred.argmax()]

        # Draw rectangle around the face and label the emotion
        cv2.rectangle(im, (x, y), (x+w, y+h), (255, 0, 0), 2)
        cv2.putText(im, prediction_label, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    # Display the output
    cv2.imshow("Facial Expression Recognition", im)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close windows
webcam.release()
cv2.destroyAllWindows()
```
